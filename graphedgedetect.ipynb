```python
import cv2,math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Get image, denoise

```python
img=cv2.imread("/Users/vigneshm/Documents/pictures/pentagon-graph.jpg")
img=cv2.GaussianBlur(img,(5,5),0)
img_bw=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
plt.imshow(img_bw,cmap='gray')
```

equalize brightness

```python
img_closed = cv2.morphologyEx(img_bw,cv2.MORPH_CLOSE,cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(19,19)))
img_bw = np.float32(img_bw)/img_closed
img_bw = cv2.normalize(img_bw,img_bw,0,255,cv2.NORM_MINMAX,cv2.CV_8U)
plt.imshow(img_bw,cmap='gray')
```

```python
edges=cv2.Canny(img_bw,0,50,5)
plt.imshow(edges,cmap='gray')
plt.show()
plt.imshow(img_bw,cmap='gray')
```

```python
contours,hier=cv2.findContours(edges,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
contours=np.array(contours)
```

```python
def draw_contour(cont,img):
    for i in range(len(cont)):
        j=(i+1)%len(cont)
        print 'line bw ',i,j
        cv2.line(img,tuple(cont[i][0]),tuple(cont[j][0]),(0,255,0))
    return img
```

```python
for contour in contours:
    contour_app=cv2.approxPolyDP(contour,cv2.arcLength(contour,True)*0.02,True)
    
    #print cv2.isContourConvex(contour_app)
    if cv2.contourArea(contour_app)<10:
        continue
#     if (not cv2.isContourConvex(contour_app)):
#         continue
    #print contour_app
    print contour_app.shape
    print cv2.contourArea(contour_app)
    plt.imshow(draw_contour(contour_app,img_bw.copy()),cmap='gray')
    plt.show()
```

```python
import cv2,math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

curve = []

def connect_points(img, point, epsilon):
    x = point[0]
    y = point[1]
    val = img[x,y]
    shape = img.shape
    
    if not ( x>1 and y>1 and x<shape[0] and y<shape[1]):
        return
    #print "curve"
    #print curve
    pts = [[x+1, y], [x-1, y], [x, y+1], [x, y-1], [x+1, y+1], [x+1, y-1], [x-1, y+1], [x-1, y-1]]
    for p in pts:
        #print p
        if not curve.count(p) == 0:
            pts.remove(p)
    
    arr = []
    for p in pts:
        arr.append(img[p[0], p[1]])
    
    #pts = np.array(pts)
    arr = np.array(arr)
    
    diff = np.fabs(arr - val)
    idx = np.argmin(diff)
    
    if diff[idx] > epsilon:
        return
    else :
        pt = pts[idx]
        curve.append(pt)
        connect_points(img, pt, epsilon)

img = cv2.imread('pentagon1.jpg', 0)
blur = cv2.GaussianBlur(img,(5,5),0)
kernel = np.ones((3,3),np.uint8)
#img = cv2.dilate(img, kernel, iterations=1)
#img = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 29, 0)
#img = cv2.erode(img, kernel, iterations=1)

connect_points(img, [109, 1226], 10)
print curve
```
